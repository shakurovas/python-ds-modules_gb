```python
X_corr = X.corr()
X_corr
```

```python
high_corr = X_corr[(abs(X_corr['target']) >= 0.5)].drop('target', axis=0)
high_corr = high_corr.index.tolist()
high_corr
```

```python
X = X.drop('target', axis=1)
```

```python
for column in X.columns:
  if column in high_corr:
    new_column_name = column + '_2'
    X[new_column_name] = X[column] ** 2
```
